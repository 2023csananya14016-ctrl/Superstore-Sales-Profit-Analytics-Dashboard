```python
#Loading Dataset
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding='latin1')
```

```python
#Renaming Column to snake_case
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.replace(' ', '_')
)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#Checking Unique Values
```

```python
for col in df.columns:
  print(col, df[col].nunique())
```

```python
#Converting Dates to proper format
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
```

```python
df
```

```python
#Creating Shipping Days
```

```python
df['shipping_days'] = (df['Ship_Date'] - df['Order_Date']).dt.days
```

```python
df
```

```python
#Removing Duplicates
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
#Making the columns standardize
```

```python
text_cols = ['Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub-Category']

for col in text_cols:
  df[col] = df[col].str.strip()
```

```python
df
```

```python
#Saving the cleaned dataset
```

```python
df.to_csv('cleaned_superstore.csv', index = False)
```

```python
#Extracting Time Features
```

```python
df['year'] = df['Order_Date'].dt.year
df['month'] = df['Order_Date'].dt.month
df['month_name'] = df['Order_Date'].dt.month_name()
df['quarter'] = df['Order_Date'].dt.quarter
df['weekday'] = df['Order_Date'].dt.day_name()
```

```python
df
```

```python
#Creating Profit Margin
```

```python
df['profit_margin'] = (df['Profit'] / df['Sales']) * 100
```

```python
df
```

```python
#Average Selling Price
```

```python
df['avg_selling_price'] = df['Sales'] / df['Quantity']
```

```python
df
```

```python
#Checking the Discount
```

```python
def discount_bucket(x):
  if x == 0:
    return 'No Discount'
  elif x <= 0.2:
    return 'Low'
  elif x <= 0.5:
    return 'Medium'
  else:
    return 'High'

df['discount_bucket'] = df['Discount'].apply(discount_bucket)
```

```python
df
```

```python
#Checking Sales
```

```python
df['sales_bucket'] = pd.qcut(df['Sales'], q=4, labels=['Low','Medium','High','Very High'])
```

```python
df
```

```python
#Customer Lifetime Value Approximation
```

```python
customer_sales = df.groupby('Customer_Name')['Sales'].transform('sum')
df['customer_lifetime_value'] = customer_sales
```

```python
df
```

```python
#Saved Engineered Dataset
```

```python
df.to_csv('engineered_superstore.csv', index=False)
```

```python
#Which region generates highest sales and profit?
```

```python
region_analysis = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
print(region_analysis)
```
